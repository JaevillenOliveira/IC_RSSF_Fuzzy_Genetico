```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

results = pd.DataFrame(index=['PacketsSent', 'PacketsSentFLC', 'PacketsReceived', 'PacketsReceivedFLC', 
                              'PacketsLost(%)', 'PacketsLostFLC(%)', 'Jitter', 'JitterFLC', 'Latency', 'LatencyFLC',
                              'PowerConsumed(W)', 'PowerConsumedFLC(W)'])

wsnSc1 = pd.read_csv('wsn.csv', usecols=['run', 'type', 'module', 'name', 'value', 'mean'])

run = "wsnSc1T"

for i in range(1, 11):
    packetSentCount = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-1')) & (wsnSc1.type=='scalar') & 
                             (wsnSc1.module.str.startswith("Sc1.sensor")) & (wsnSc1.module.str.endswith("udp")) & 
                             (wsnSc1.name=='packetSent:count')]   

    packetReceivedCount = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-1'))& (wsnSc1.type=='scalar') & 
                                 (wsnSc1.module.str.startswith("Sc1.sink")) & 
                                 (wsnSc1.module.str.endswith("udp")) & (wsnSc1.name=='packetReceived:count')]

    totalPkSentCount = sum(packetSentCount.value)
    totalPkReceivedCount = sum(packetReceivedCount.value)
    packetLoss = totalPkSentCount - totalPkReceivedCount

    residualEnergy = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-1')) & (wsnSc1.type=='scalar') & 
                            (wsnSc1.name=='residualEnergyCapacity:last') & 
                            (wsnSc1.module.str.endswith('energyStorage'))] 

    spentEnergyJ = 3000 - sum(residualEnergy.value)
    consumedPower = spentEnergyJ / 240

    latency = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-1')) & (wsnSc1.type=='histogram') & 
                     (wsnSc1.name=='endToEndDelay:histogram')] 

    latencyMean = latency["mean"].mean()
    
    jitter = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-1')) & (wsnSc1.type=='histogram') & 
                    (wsnSc1.name=='jitter:histogram')] 
    
    jitterMean = jitter["mean"].mean()
    
                        ########################################################
                        ########Simulation with FUZZY LOGIC CONTROLLER##########
                        ########################################################
    
    packetSentCountFLC = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-0')) & (wsnSc1.type=='scalar') & 
                                (wsnSc1.module.str.startswith("Sc1.sensor")) & 
                                (wsnSc1.module.str.endswith("udp")) & (wsnSc1.name=='packetSent:count')] 

    packetReceivedCountFLC = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-0'))& (wsnSc1.type=='scalar') & 
                                    (wsnSc1.module.str.startswith("Sc1.sink")) & 
                                    (wsnSc1.module.str.endswith("udp")) & (wsnSc1.name=='packetReceived:count')]

    totalPkSentCountFLC = sum(packetSentCountFLC.value)
    totalPkReceivedCountFLC = sum(packetReceivedCountFLC.value)
    packetLossFLC = totalPkSentCountFLC - totalPkReceivedCountFLC

    residualEnergyFLC = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-0')) & (wsnSc1.type=='scalar') & 
                               (wsnSc1.name=='residualEnergyCapacity:last') & 
                               (wsnSc1.module.str.endswith('energyStorage'))] 

    spentEnergyJFLC = 3000 - sum(residualEnergyFLC.value)
    consumedPowerFLC = spentEnergyJFLC / 240

    latencyFLC = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-0')) & (wsnSc1.type=='histogram') & 
                        (wsnSc1.name=='endToEndDelay:histogram')] 

    latencyMeanFLC = latencyFLC["mean"].mean()

    jitterFLC = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-0')) & 
                       (wsnSc1.type=='histogram') & (wsnSc1.name=='jitter:histogram')] 

    jitterMeanFLC = jitterFLC["mean"].mean()   
    
    handoverFLC = wsnSc1[(wsnSc1.run.str.startswith('wsnSc1T'+str(i)+'-0')) & (wsnSc1.type=='scalar') & 
                    (wsnSc1.name=='handoverTime:mean')] 
    
    handoverFLC = handoverFLC["value"]
    
                        ########################################################
                        ########                 RESULTS             ##########
                        ########################################################
            
    runDataCollector = [totalPkSentCount, totalPkSentCountFLC, totalPkReceivedCount, totalPkReceivedCountFLC,
                        packetLoss, packetLossFLC, jitterMean, jitterMeanFLC, latencyMean, latencyMeanFLC, 
                        consumedPower, consumedPowerFLC]
    results.insert(i-1, "T"+str(i), runDataCollector)
        
    print(handoverFLC)
```
